```python
from PIL import Image, ImageFilter 
  
image = Image.open(r"comida.jpg") 

# Las dimensiones del kernel de ImageFilter.GaussianBlur son 5 × 5. Al final mostramos la imagen.
# GaussianBlur(x) x => radio de desenfoque o desviacion estandar(tamaño del kernel en una dirección, desde el píxel central)

image = image.filter(ImageFilter.GaussianBlur(2)) 
  
display(image)
```

```python
image = image.filter(ImageFilter.GaussianBlur(8)) 
  
display(image)
```

```python
image = image.filter(ImageFilter.GaussianBlur(15)) 
  
display(image)
```

```python
tamaño = (3,3)
 
coeficientes = [1, 2, 1, 2, 4, 2, 1, 2, 1]

imagen_procesada = image.filter(ImageFilter.Kernel(tamaño, coeficientes))

display(imagen_procesada)
```

```python
tamaño = (5,5)
 
coeficientes = [0, 1, 2, 1,0, 1, 3, 5, 3, 1, 2, 5, 9, 5, 2, 1, 3, 5, 3, 1, 0, 1, 2, 1, 0]

imagen_procesada = image.filter(ImageFilter.Kernel(tamaño, coeficientes))

display(imagen_procesada)
```
