```python
#Actividad 5:
def sumacomplejos(lista_numeroscomplejos):
    lista_numeroscomplejos[2][0]=lista_numeroscomplejos[0][0]+lista_numeroscomplejos[1][0]
    lista_numeroscomplejos[2][1]=lista_numeroscomplejos[0][1]+lista_numeroscomplejos[1][1]
def main():
    print("Este programa realiza la suma de 2 números complejos")
    lista_numeroscomplejos=[[None,None],[None,None],[None,None]]
    for i in range(2):
        lista_numeroscomplejos [i][0]= int(input("Introduce la parte real del numero {}:  ".format(i+1)))
        lista_numeroscomplejos [i][1]= int(input("Introduce la parte compleja del numero {}:  ".format(i+1)))
    print("El primer número complejo es {}+{}i".format(lista_numeroscomplejos[0][0],lista_numeroscomplejos[0][1]))
    print("El segundo número complejo es {}+{}i".format(lista_numeroscomplejos[1][0],lista_numeroscomplejos[1][1]))
    sumacomplejos(lista_numeroscomplejos)
    print("El resultado de la suma es {}+{}i".format(lista_numeroscomplejos[2][0],lista_numeroscomplejos[2][1])) 
main()
```

```python
#Actividad 6:
def descuento(numcamisas,preciocamisa):
    if numcamisas>=3:
        descuent=.2
    else:
        descuent= .1
    preciototal= preciocamisa*numcamisas
    preciototaldescuento=preciototal-(preciototal*descuent)
    return preciototaldescuento
def main():    
    print("Este programa calcula el precio total de las camisas aplicando un descuento")
    numcamisas=int(input("Introduce el número de camisas a comprar $"))
    preciocamisa=int(input("Introduce el precio de las camisas a comprar "))
    if numcamisas<0:
        print("Número invalido de camisas")
        return 0
    preciototaldescuento=descuento(numcamisas,preciocamisa)
    print("El precio total de las camisas aplicando un descuento es de ${}".format(preciototaldescuento))
main()
```

```python
#Actividad 7:
def utilidad(paga,antiguedadt):
    if antiguedadt>=120:
        utilidadanual=paga*.2
    else:
        if antiguedadt>=60:
            utilidadanual=paga*.15
        else:
            if antiguedadt>=24:
                utilidadanual=paga*.1
            else:
                if antiguedadt>=12:
                    utilidadanual=paga*.07
                else:
                    utilidadanual=paga*.05
    return utilidadanual
def main():    
    print("Este programa calcula la utilidad que un trabajador que recibe en el reparto anual de utilidades ")
    antiguedadaño=int(input("Introduzca la antiguedad del empleado\n años: "))
    antiguedadmes=int(input("meses: "))
    antiguedadt=antiguedadaño*12 + antiguedadmes
    print("La antiguedad del empleado es de {} año(s) y {} mes(es) que equivalen a {} meses".format(antiguedadaño,antiguedadmes,antiguedadt))
    paga=int(input("Introduzca el pago mensual que recibe: $"))
    if paga<0:
        print("Número invalido de pago mensual")
        return 0
    if antiguedadt<0:
        print("Número invalido en pago mensual")
        return 0
    utilidadanual=utilidad(paga,antiguedadt)
    print("La utilidad anual del empleado es de ${}".format(utilidadanual))
main()
```

```python
##### Actividad 9:
>>> import math
def Seno(grados):
    res=math.sin(math.radians(grados))
    return res
def Coseno(grados):
    res=math.cos(math.radians(grados))
    return res
def Tangente(grados):
    res=math.tan(math.radians(grados))
    return res
def Cotangente(radianes):
    res=math.degrees(math.atan(math.radians(radianes)))
    return res
def Secante(radianes):
    res=math.degrees(math.acos(math.radians(radianes)))
    return res
def Cosecante(radianes):
    res=math.degrees(math.asin(radianes))
    return res
def main():    
    print("Calculadora de operaciones trigonométricas,permite calcular el seno, coseno, tangente y sus inversas.")
    while True:
        print("\nQue deseas hacer?\n1)Realizar seno\n2)Realizar coseno")
        print("3)Realizar tangente\n4)Realizar cotangente\n5)Realizar secante")
        print("6)Realizar cosecante\n7)Salir")
        op=int(input())
        if op<=3:
            grados=int(input("Ingrese los grados para efectuar la operación: "))
            if op==1:
                res=Seno(grados)
            elif op==2:
                res=Coseno(grados)
            elif op==3:
                res=Tangente(grados)
            print("El resultado de la operación es: {} radianes".format(res))
        elif 3<op<=6:
            radianes=int(input("Ingrese los radianes para efectuar la operación: "))
            if op==4:
                res=Cotangente(radianes)
            elif op==5:
                res=Secante(radianes)
            elif op==6:
                res=Cosecante(radianes)
            print("El resultado de la operación es: {} grados".format(res))
        elif op==7:
                break
main()
```
